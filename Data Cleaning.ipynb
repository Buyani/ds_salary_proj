```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df=pd.read_csv('glassdoor_jobs.csv')
df.head(3)
```

#### Tasks

#salary parsing
#Compnay name text only
#state field
#age of a company
#parsing job description (python etc.)

```python
df=df[df['Salary Estimate']!='-1']
```

```python
salary=df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary.head(10)
```

```python
minus_kd=salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
minus_kd.head()
```

```python
#Make salary for per hour
df["Hourly"]=df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df['employer_provided']=df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
df.head()
```

```python
minus_kd.head()
```

```python
remove_per_hour=minus_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
remove_per_hour.head()
```

```python
remove_per_hour=remove_per_hour.apply(lambda x: x.replace(':',''))
```

```python
df['min_salary']=remove_per_hour.apply(lambda x: x.split('-')[0])
df['max_salary']=remove_per_hour.apply(lambda x: x.split('-')[1])
```

```python
df.head(5)
```

```python
df['max_salary']=pd.to_numeric(df['max_salary'])
df['min_salary']=pd.to_numeric(df['min_salary'])
```

```python
df.head()
```

```python
df['avg_salary']=(df.min_salary+df.max_salary)/2
```

```python
df.head()
```

```python
#Persing Company Name
df['company']=df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3],axis=1)
```

```python
df.head()
```

```python
df['job_state']=df['Location'].apply(lambda x:x.split(',')[1])
```

```python
df.job_state.value_counts()
```

```python
df['Age']=df.Founded.apply(lambda x: x if x < 1 else 2021-x)
```

```python
df.head()
```

```python
df['same_state']=df.apply(lambda x: 1 if x.Location==x.Headquarters else 0,axis=1)
```

```python
df.head()
```

```python
#age of company 
df['age'] = df.Founded.apply(lambda x: x if x <1 else 2020 - x)

#parsing of job description (python, etc.)

#python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
 
#r studio 
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R_yn.value_counts()

#spark 
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark.value_counts()

#aws 
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.aws.value_counts()

#excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel.value_counts()

df.columns

df_out = df.drop(['Unnamed: 0'], axis =1)

df_out.to_csv('salary_data_cleaned.csv',index = False)
```
